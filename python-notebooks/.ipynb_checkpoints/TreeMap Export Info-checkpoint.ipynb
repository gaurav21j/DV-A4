```python
import numpy as np 
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
#global str
import json
```

```python
export_india = pd.read_csv('../dataset/2018-2010_export.csv')
```

```python
year_list = list(export_india.year.unique())
print(year_list)
```

```python
data_18 = export_india[export_india['year']==2018]
commodity_list = list(data_18.Commodity.unique())
commodity_group = data_18.groupby('Commodity')
ls1 = []

for year in year_list:

    ls = []
    for commodity_name in commodity_list:
        ls.append([commodity_name, (commodity_group.get_group(commodity_name).value.sum())/1000 ])

    total = pd.DataFrame(ls, columns = ['Commodity', 'total_exports'])     
    total = total.sort_values(by='total_exports', ascending=False)
    total = total.head(4)

    #Now Getting Country Information
    country_list = list(data_18.Commodity.unique())
    country_group = data_18.groupby('Commodity')

    for index, row in total.iterrows():
        data_com_18 = data_18[data_18['Commodity'] == row['Commodity']]
        data_com_18 = pd.DataFrame(data_com_18, columns = ['country', 'value'])
        data_com_18 = data_com_18.sort_values(by='value', ascending=False)
        data_com_18 = data_com_18.head(4)

        for index, row1 in data_com_18.iterrows():
            data = {'key':row1['country'], 'region':year, 'subregion':row['Commodity'], 'value':row1['value']}
            ls1.append(data)

```
